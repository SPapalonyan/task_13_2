```
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

def find_keywords(file_path, text_column='normalized_text', top_n=5):
    df = pd.read_csv(file_path)

    if text_column not in df.columns:
        raise ValueError(f"В файле должен быть столбец '{text_column}'")

    texts = df[text_column].fillna('')

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(texts)

    feature_names = vectorizer.get_feature_names_out()

    keywords_list = []

    for i in range(tfidf_matrix.shape[0]):
        row = tfidf_matrix[i].toarray()[0]
        top_indices = row.argsort()[-top_n:][::-1]

        keywords = [
            (feature_names[idx], float(row[idx]))
            for idx in top_indices if row[idx] > 0
        ]

        keywords_list.append(keywords)

    df['key_words'] = keywords_list

    return df

url = "https://raw.githubusercontent.com/SPapalonyan/task_13_2/refs/heads/main/cafe_reviews.csv"

df = find_keywords(url, text_column='normalized_text', top_n=5)

print(df.head())
```
